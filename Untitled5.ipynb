```python
from pathlib import Path
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

# 1. 경로 설정
DATA_DIR = Path('/content/type3')
file_path = DATA_DIR / 'channel_purchase.csv'

# 2. 데이터 불러오기 (파일이 없으면 자동으로 테스트용 데이터 생성)
try:
    df = pd.read_csv(file_path)
    print("📁 [안내] 업로드된 실제 데이터를 성공적으로 불러왔습니다.")
except FileNotFoundError:
    print("⚠️ [안내] 경로에 파일이 없어 테스트용 가상 데이터를 자동 생성하여 분석을 진행합니다.")
    # 실제 파일이 없을 때 오류를 방지하기 위한 가상 데이터 생성 로직
    np.random.seed(42)
    df = pd.DataFrame({
        'channel': np.random.choice(['Search', 'Social', 'Direct', 'Organic'], 800, p=[0.4, 0.3, 0.2, 0.1]),
        'purchase_yn': np.random.choice([0, 1], 800, p=[0.75, 0.25])
    })

# 3. 데이터 확인
print('\n[데이터 상위 5행]')
print(df.head(), '\n')

print('[변수 정보]')
df.info()
print('\n')

# 4. 교차표 생성
table = pd.crosstab(df['channel'], df['purchase_yn'])
print('[교차표]')
print(table, '\n')

# 5. 카이제곱 독립성 검정
result = chi2_contingency(table)
chi2 = result[0]
p = result[1]
dof = result[2]
expected = result[3]

print('[검정 결과]')
print('chi-square statistic:', round(chi2, 4))
print('p-value:', round(p, 6) if p >= 0.0001 else f"{p:.6e}")
print('degrees of freedom:', dof)

# 6. 기대도수 확인
expected_df = pd.DataFrame(expected, index=table.index, columns=table.columns)
print('\n[기대도수]')
print(expected_df.round(2))

# 7. 채널별 구매전환율 계산
if df['purchase_yn'].dtype == 'object':
    is_purchased = df['purchase_yn'].isin(['Y', 'yes', '1', 'True'])
else:
    is_purchased = df['purchase_yn'] == 1

conversion_rate = df.groupby('channel').apply(lambda x: (is_purchased[x.index].mean() * 100), include_groups=False).sort_values(ascending=False)

print('\n[채널별 구매전환율(%)]')
print(conversion_rate.round(2))

# 8. 해석 출력
print('\n[해석]')
if p < 0.05:
    print('p-value가 0.05보다 작으므로, 유입채널과 구매여부는 독립이 아니며 서로 관련이 있다고 해석할 수 있습니다.')
else:
    print('p-value가 0.05 이상이므로, 유입채널과 구매여부의 관련성을 확인하기 어렵습니다.')

if not conversion_rate.empty:
    print(f'가장 높은 구매전환율 채널은 [{conversion_rate.index[0]}]이며, 전환율은 {conversion_rate.iloc[0]:.2f}%입니다.')
```

```python
from pathlib import Path
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

# 경로 설정
DATA_DIR = Path('/content/type3')
file_path = DATA_DIR / 'gender_preference.csv'

# 1. 데이터 불러오기 (파일이 없으면 자동으로 테스트용 데이터 생성)
try:
    df = pd.read_csv(file_path)
    print("📁 [안내] 업로드된 실제 데이터를 성공적으로 불러왔습니다.")
except FileNotFoundError:
    print("⚠️ [안내] 경로에 파일이 없어 테스트용 가상 데이터를 자동 생성하여 분석을 진행합니다.")
    # 실제 파일이 없을 때 오류를 방지하기 위한 가상 데이터 생성 로직
    np.random.seed(42)
    df = pd.DataFrame({
        'gender': np.random.choice(['Male', 'Female'], 500, p=[0.45, 0.55]),
        'content_type': np.random.choice(['Drama', 'Action', 'Comedy', 'Sci-Fi'], 500, p=[0.4, 0.2, 0.3, 0.1])
    })

# 2. 데이터 확인
print('\n[데이터 상위 5행]')
print(df.head(), '\n')

print('[변수 정보]')
df.info()
print('\n')

# 3. 교차표 생성
table = pd.crosstab(df['gender'], df['content_type'])

print('[교차표]')
print(table, '\n')

# 4. 카이제곱 동질성 검정
result = chi2_contingency(table)
chi2 = result[0]
p = result[1]
dof = result[2]
expected = result[3]

print('[검정 결과]')
print('chi-square statistic:', round(chi2, 4))
print('p-value:', round(p, 6) if p >= 0.0001 else f"{p:.6e}")
print('degrees of freedom:', dof)

# 5. 기대도수 확인
expected_df = pd.DataFrame(expected, index=table.index, columns=table.columns)

print('\n[기대도수]')
print(expected_df.round(2))

# 6. 성별별 최다 선호 유형 확인
top_pref = table.idxmax(axis=1)

print('\n[성별별 최다 선호 유형]')
print(top_pref)

# 7. 행 비율(성별 내 비율) 확인
row_ratio = pd.crosstab(df['gender'], df['content_type'], normalize='index') * 100

print('\n[성별 내 콘텐츠 선호 비율(%)]')
print(row_ratio.round(2))

# 8. 해석 출력
print('\n[해석]')
if p < 0.05:
    print('p-value가 0.05보다 작으므로, 성별에 따라 콘텐츠 선호 분포가 다르다고 해석할 수 있습니다. (귀무가설 기각)')
else:
    print('p-value가 0.05 이상이므로, 성별에 따른 콘텐츠 선호 분포 차이를 확인하기 어렵습니다. (귀무가설 채택)')

for g in top_pref.index:
    print(f'{g}의 최다 선호 콘텐츠는 [{top_pref[g]}]입니다.')
```
