```python
import pandas as pd
from scipy import stats
from pathlib import Path
import os

# 1. 드라이브 마운트 (직접 마운트 명령 사용)
# 만약 에러가 계속되면 왼쪽 폴더 아이콘을 눌러 수동으로 마운트하세요.
from google.colab import drive
try:
    if not os.path.exists('/content/drive'):
        drive.mount('/content/drive', force_remount=True)
    print("구글 드라이브 마운트 성공")
except Exception as e:
    print(f"드라이브 마운트 실패: {e}")

# 2. 경로 설정 및 파일 로드
# 정확한 경로인지 확인하기 위해 절대 경로를 문자열로 직접 지정하는 것이 안전할 수 있습니다.
FILE_PATH = '/content/drive/MyDrive/type3_week1/branch_satisfaction.csv'

if os.path.exists(FILE_PATH):
    df = pd.read_csv(FILE_PATH)

    # 3. 데이터 확인 및 결측치 제거
    print('[데이터 미리보기]')
    print(df.head(), '\n')

    # 분석 대상 데이터 분리 (결측치 제거 포함)
    groups = []
    for branch in ['A', 'B', 'C']:
        data = df[df['branch'] == branch]['score'].dropna()
        groups.append(data)

    # 4. 일원분산분석 (ANOVA)
    result = stats.f_oneway(*groups)
    means = df.groupby('branch')['score'].mean().sort_values(ascending=False)

    # 5. 결과 출력
    print('[분석 결과]')
    print('지점별 평균:')
    print(means, '\n')
    print(f'F-statistic: {round(result.statistic, 4)}')
    print(f'p-value: {round(result.pvalue, 6)}')
    print(f'평균이 가장 높은 지점: {means.index[0]}')

    print('\n[결과 해석]')
    if result.pvalue < 0.05:
        print('p-value가 0.05보다 작으므로 지점 간 평균 고객만족도 차이가 통계적으로 유의미함.')
    else:
        print('p-value가 0.05보다 크므로 지점 간 평균 고객만족도 차이가 유의하다고 보기 어려움.')
else:
    print(f"파일을 찾을 수 없습니다. 경로를 다시 확인하세요: {FILE_PATH}")
```

```python
from google.colab import drive
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
from statsmodels.stats.anova import anova_lm
from pathlib import Path
import os

# 1. 드라이브 마운트 확인 및 수행
if not os.path.exists('/content/drive'):
    try:
        drive.mount('/content/drive')
    except:
        print("드라이브 마운트에 실패했습니다. 왼쪽 폴더 아이콘을 눌러 수동 마운트를 시도하세요.")

# 2. 경로 설정 및 파일 로드
DATA_DIR = Path('/content/drive/MyDrive/type3_week1')
file_path = DATA_DIR / 'ad_age_sales.csv'

if file_path.exists():
    df = pd.read_csv(file_path)

    # 3. 데이터 미리보기
    print('[데이터 미리보기]')
    print(df.head(), '\n')

    # 4. 이원분산분석(Two-way ANOVA) 모델 구성
    # sales_amount: 종속변수
    # ad_type, age_group: 독립변수 (범주형)
    model = smf.ols('sales_amount ~ C(ad_type) * C(age_group)', data=df).fit()

    # typ=2는 상호작용 항이 있는 경우 권장되는 계산 방식입니다.
    anova_table = anova_lm(model, typ=2)

    print('[ANOVA 결과표]')
    print(anova_table, '\n')

    # 5. 주요 p-value 추출
    p_ad = anova_table.loc['C(ad_type)', 'PR(>F)']
    p_age = anova_table.loc['C(age_group)', 'PR(>F)']
    p_inter = anova_table.loc['C(ad_type):C(age_group)', 'PR(>F)']

    print('[핵심 p-value]')
    print(f'광고유형 p-value: {p_ad:.6f}')
    print(f'연령대 p-value: {p_age:.6f}')
    print(f'상호작용 p-value: {p_inter:.6f}')

    # 6. 결과 해석
    print('\n[결과 해석]')
    if p_inter < 0.05:
        print('상호작용 효과가 유의하므로, 광고유형의 효과는 연령대에 따라 달라진다고 해석함.')
    else:
        print('상호작용 효과가 유의하지 않으므로, 광고유형과 연령대의 주효과 중심으로 해석함.')
        if p_ad < 0.05:
            print('- 광고유형에 따른 매출 차이는 유의미함.')
        if p_age < 0.05:
            print('- 연령대에 따른 매출 차이는 유의미함.')
else:
    print(f"파일을 찾을 수 없습니다: {file_path}")
```
